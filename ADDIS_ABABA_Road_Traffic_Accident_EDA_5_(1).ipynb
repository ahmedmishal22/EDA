#Descrpition

Time: In which time the accident have been done

Day_of_week: Accident day

Age_band_of_driver: Age of the driver

Sex_of_driver: The drivers gender (Male/Female)

Educational_level: Driver qualification

Vehicle_driver_relation: Relation between the driver and the vehicle

Driving_experience: How long the driver has been roat the vehicle

Type_of_vehicle: vehicle type (petrol/disel/automatic)

Owner_of_vehicle: governmet vehicle or personal

Service_year_of_vehicle: How long the vehicle has been serviced

Vehicle_movement: movement of the vehicle

Casualty_class: it  is used in the context of road traffic accidents to categorize the involvement of individuals in an accident. It helps in analyzing accident data and understanding the different types of casualties involved.

Sex_of_casualty: it refers to the biological sex of a person involved in a road traffic accident. This information is often collected as part of accident data

Age_band_of_casualty: it refers to the age group of a person involved in a road traffic accident. This information is often collected as part of accident data and used for various purposes

Casualty_severity: classification system used in road traffic accident data to categorize the seriousness of injuries sustained by individuals involved in an accident.

Work_of_casuality: data field used in road traffic accident analysis to categorize the occupation or employment status of individuals involved in an accident.

Fitness_of_casuality: data field used in road traffic accident analysis to indicate the physical and mental condition of an individual involved in an accident.

pedestrian_movement: refers to the patterns and behaviors of individuals walking in various environments, including urban areas, parks, and public spaces


Cause_of_accident: it  is a critical factor in understanding road traffic accidents. It helps identify patterns, implement preventive measures, and improve overall road safety.

Accident_severity: is a classification system used to categorize the seriousness of injuries and fatalities resulting from a road traffic accident.

```python
import pandas as pd
# Example: Assume your data is already loaded in 'df'
# Convert the Time column to datetime format
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce')

# Define the time ranges for day and night
day_start = pd.to_datetime('06:00:00', format='%H:%M:%S').time()  # 6 AM
day_end = pd.to_datetime('18:00:00', format='%H:%M:%S').time()    # 6 PM

# Create a new column to classify as 'Day' or 'Night'
df['Day_or_Night'] = df['Time'].apply(lambda x: 'Day' if day_start <= x.time() <= day_end else 'Night')

# Check the result
print(df[['Time', 'Day_or_Night']])
```

in this step we sort the time and convert it into day and night

```python
# plot the count plot of the accidents by time
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Day_or_Night', palette='rainbow')
plt.title('Count of Accidents', fontsize=16)
plt.xlabel('Time of Day', fontsize=12)
plt.ylabel('Count of accident', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
# take the value count of accident severity
value_counts = df['Accident_severity'].value_counts()
print(value_counts)
```

```python
# Extract labels and sizes for the pie chart
labels = value_counts.index.tolist()  # Get unique severity levels as labels
sizes = value_counts.values.tolist()  # Get corresponding counts as sizes

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
# autopct formats percentages, startangle rotates the chart

plt.title("Accident Severity Distribution")
plt.axis('equal')  # Ensure a circular pie chart
plt.show()
```

# Data Cleaning

cleaning the above dataset

```python
# dropping columns that can cause imbalance while imputation
lists = ['Vehicle_driver_relation', 'Work_of_casuality', 'Fitness_of_casuality',
         'Day_of_week', 'Casualty_severity', 'Time', 'Sex_of_driver',
         'Educational_level', 'Defect_of_vehicle', 'Owner_of_vehicle',
         'Service_year_of_vehicle', 'Road_surface_type', 'Sex_of_casualty']

df.drop(columns=lists, inplace=True, errors='ignore')
```

```python
# fill missing values with mean column values
df['Driving_experience'].fillna(df['Driving_experience'].mode()[0], inplace=True)
df['Age_band_of_driver'].fillna(df['Age_band_of_driver'].mode()[0], inplace=True)
df['Type_of_vehicle'].fillna(df['Type_of_vehicle'].mode()[0], inplace=True)
df['Area_accident_occured'].fillna(df['Area_accident_occured'].mode()[0], inplace=True)
df['Road_allignment'].fillna(df['Road_allignment'].mode()[0], inplace=True)
df['Type_of_collision'].fillna(df['Type_of_collision'].mode()[0], inplace=True)
df['Vehicle_movement'].fillna(df['Vehicle_movement'].mode()[0], inplace=True)
df['Lanes_or_Medians'].fillna(df['Lanes_or_Medians'].mode()[0], inplace=True)
df['Types_of_Junction'].fillna(df['Types_of_Junction'].mode()[0], inplace=True)
```

```python
# remove the unwanted items
df['Driving_experience'] = df['Driving_experience'].str.replace("yr","year")
df['Driving_experience'] = df['Driving_experience'].str.replace("Above"," ")
```

```python
df['Type_of_vehicle'] = df['Type_of_vehicle'].str.replace("("," ")
df['Type_of_vehicle'] = df['Type_of_vehicle'].str.replace(">"," ")
df['Type_of_vehicle'] = df['Type_of_vehicle'].str.replace(")"," ")
df['Type_of_vehicle'] = df['Type_of_vehicle'].str.replace("Public","Public:")
df['Type_of_vehicle'] = df['Type_of_vehicle'].str.replace("41?100Q"," ")
```

```python
df['Age_band_of_driver'] = df['Age_band_of_driver'].str.replace("Unknown"," ")
```

```python
df['Lanes_or_Medians'] = df['Lanes_or_Medians'].str.replace("(divided with broken lines road marking)"," ")
```

```python
df['Age_band_of_driver'] = df['Age_band_of_driver'].str.replace("Undivided"," ")
```

```python
df['Light_conditions'] = df['Light_conditions'].str.replace("- lights lit	"," ")
```
